```python
#understanding the connectome matrix
connectome_data = pd.read_csv("/Users/folasewaabdulsalam/Health-Tech-Projects/ADHD-Kaggle/TRAIN/TRAIN_FUNCTIONAL_CONNECTOME_MATRICES.csv")
connectome_data.head(5)
```

```python
participant_id = connectome_data['participant_id']

connectome_values = connectome_data.drop(columns = ['participant_id']).to_numpy() #dropping the id column

num_regions = 200 #200x200 matrix
num_connections = num_regions * (num_regions-1) // 2 #upper triangular elements

#reconstructing the data into a full connectivity matrix

def reconstruct_connectome_vector(connectome_vector, num_regions):
    """
    A function to convert the flattened upper traingular vector to a full matrix
    """
    full_matrix = np.zeros((num_regions, num_regions))
    triu_indices = np.triu_indices(num_regions, k=1) #returns the indices of the upper triangular part of a square matrix

    full_matrix[triu_indices] = connectome_vector #filling up the upper triangle

    full_matrix += full_matrix.T

    return full_matrix

#converting the row back to a 200x200 matrix 

connectome_matrix = np.array([reconstruct_connectome_vector(row, num_regions) for row in connectome_values])

print("Reconstructed connectome shape: ", connectome_matrix.shape)
```

```python
#visualizing a sample connectome matrix
random_idx = np.random.randint(len(connectome_matrix))
plt.figure(figsize = (10, 8))
sns.heatmap(connectome_matrix[random_idx], cmap='coolwarm',center=0)
plt.title(f"Functional Connectome Matrix for Participant {participant_id[random_idx]} ")
plt.show()
```

```python
#merging the data
metadata = pd.merge(categorical_metadata, quantitative_data, on = ['participant_id'], how = 'inner')
metadata_with_labels = pd.merge(metadata, training_labels, on=['participant_id'], how = 'inner')
train_data = pd.merge(metadata_with_labels, connectome_data, on = ['participant_id'], how = 'inner')
train_data.head(5)
```
